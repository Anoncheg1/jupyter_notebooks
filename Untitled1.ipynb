# Зеленая зона выданных кредитов

```python
import pandas as pd
from evseeva import feature_importance_forest
p = 'encoded.pickle'
df: pd.DataFrame = pd.read_pickle(p)
df.loc[(df['СФ андерайтера'] == 0) & (df['Коды отказа'] != 97) & (df['Коды отказа'] != 91), "flag"] = True
df['flag'].fillna(False, inplace=True)
df.drop('СФ андерайтера', axis=1, inplace=True)
df.drop('Коды отказа', axis=1, inplace=True)
feature_importance_forest(df)
```
